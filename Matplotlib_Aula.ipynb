# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g')

plt.subplot(1,2,2)
plt.plot(y,x,'y')
```

### MÃ©todo Orientado a Objetos

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])


axes1.plot(x,y,'purple')
axes1.set_title('LARGER PLOT')

axes2.plot(y,x,'orange')
axes2.set_title("SMALL PLOT")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x,'red')
axes[1].set_title('Second Plot')
```

### Figure Size and DPI

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x,'red')

plt.tight_layout()
```

```python
fig.savefig('my_picture.PDF', dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label='X Squared', marker='2',markersize='20')
ax.plot(x,x**3, label='X Cubed', marker='o', markerfacecolor='green')


ax.legend(loc=0)
```
